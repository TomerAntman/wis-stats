# Censored and truncated distributions (63)

Say experimenters A and B have a behavioral assay in which they time how long it takes for a mouse to descend a pole. Experimenter A records the descent times as follows.

- If the mouse descends the pole in less than $y_{max}$ seconds (typically 60 seconds), he records the descent time.
- If the mouse does not descend the pole in less than $y_{max}$ seconds, he does not record anything. He simply ignores that trial.

Experimenter B takes a different approach. She records the descent times as follows.

- If the mouse descends the pole in less than $y_{max}$ seconds, she records the descent time.
- If the mouse does not descend the pole in less than $y_{max}$ seconds, she records the descent time as $y_{max}$ seconds.

Let $y$ be the time to descend the pole for patient experimenters. That is, $y$ can be greater than $y_{max}$ seconds. Let $f(y)$ by the probability density function of descent time and $F(y)$ be the corresponding CDF.

**a)** Write down an expression for $f_\text{trunc}(y)$, the probability density function for experimenter A's observations in terms of $f(y)$ and $F(y)$. This is called a *truncated* distribution.

**b)** Write down an expression for $f_\text{cens}(y)$, the probability density function for experimenter B's observations that are *not* $y_{max}$. Also write an expression for $\pi(y_{max})$, which is the probability that experimenter B records $y_{max}$. This is called a *censored* distribution.

# Answers
--------------------------------------------------------------------------
Let $f(y)$ be the PDF of descent time and $F(y)$ be the corresponding CDF.<br>
## A
Let $f_{trunc}(y)$ be the PDF of experimenter A and let all recorded $y$ values be less than $y_{max}$.<br>
Since the values are naturally limited by 0 we can claim that $F(y=0)=0$.<br>
Hence we can say that the experiment in these terms creates a **right-truncated distribution**. So according to the conditional PDF function we can say that:
$$
f(y|y < y_{max})=\frac{f(y)}{F(y_{max})} 
$$

Since there was no recording if $y\ge y_{max}$ and $y$ isn't defined below 0, we can say that:
$$
f_{\text{trunc}}(y)=\frac{f(y)}{F(y_{max})} \quad \text{if } 0 \le y < y_{max} 
$$


Or:
$$
f_{\text{trunc}}(y) = \begin{cases}
    \frac{f(y)}{F_A(y_{max})} & \text{if } 0 \le y < y_{max} \\
    N/A & \text{otherwise}
\end{cases}
$$

In "truncation" we are limited by the conditions to only observe values below $y_{max}$ and therefore are required to normlize the density function.

Let's make sure the probability summs to 1 (assuming measurements are bound between 0 and $y_{max}$ and that $F(0)=0$):
$$
P(0 \le y \le y_{max})=\int_{-\infty}^{\infty}f(y)\,dy =\int_{0}^{y_{max}} f_{\text{trunc}}(y) \, dy = \int_{0}^{y_{{max}}} \frac{f(y)}{F(y_{{max}})} \, dy \\

= \frac{1}{F(y_{max})} \int_{0}^{y_{{max}}} f(y) \, dy = \frac{1}{F(y_{max})} \cdot \left[F(y_{max}) - F(0)\right] \\

= \frac{1}{F(y_{max})} \cdot F(y_{{max}}) = 1
$$

-----------------------------------
## B
Let $f_\text{cens}(y)$ be the PDF for experimenter B ($B$) and let the maximal recorded value of $y$ be $y_{max}$. <br>
As long as the observations are continuous ($y < y_{max}$) the PDF $f_\text{cens}(y)$ is identical to the original PDF. <br>
This can be simply written as:
$$
f(y|y<y_{max})=f(y)
$$
Or:
$$
f_\text{cens}(y) = f(y) \quad \text{for } y < y_{max}
$$

But we have a special case for $y \le y_{max}$ (they are not ignored like in exp'A conditions).<br>
The probability that exp'B records exactly $y_{max}$ is the probability of $y \ge y_{max}$:
$$
\pi(y_{max}) = P(y \ge y_{max}) = 1 - P(y < y_{max})
$$
So the probability that exp'B recorded $y_{max}$ is:
$$
\pi(y_{max}) = 1 - F(y_{max})
$$

In "Censoring" we create a mixed distribution where the continuous part remains **unchanged below the threshold**, but all probability mass above the threshold get concenctrated exactly at the threshold.

We can make sure and see that the total probabilty is 1 (assuming measurements are bound between 0 and $y_{max}$ and that $F(0)=0$):
$$
P(0 \le y \le y_{max})=\int_{-\infty}^{\infty}f(y)\,dy = \int_{-\infty}^{\infty} f_{\text{cens}}(y)\,dy + \pi(y_{max})= \\
=\int_{0}^{y_{max}} f_{\text{cens}}(y)\,dy + \pi(y_{max}) = \int_{0}^{y_{max}} f(y) \, dy + [1 - F(y_{max})] \\
= \left( F(y_{max}) - F(0) \right) + 1 - F(y_{max}) = 1
$$
